```python
import pandas as pd

df = pd.read_csv('Istanbul Weather Data.csv', sep=',')
df.head() #show the first 5 rows
```

```python
df = df[["Condition", "MinTemp", "AvgHumidity"]]
df.head()
```

```python
df.shape #print the dimensions of this dframe
```

```python
df = df.dropna() #remove empty rows

#Convert each category into an interger
for ind in df.index:
  if df["Condition"][ind]=='Sunny':
    df["Condition"][ind] = 0
  elif df["Condition"][ind]=='Partly cloudy':
    df["Condition"][ind] = 1
  elif df["Condition"][ind]=='Cloudy':
    df["Condition"][ind] = 2
  elif df["Condition"][ind]=='Overcast':
    df["Condition"][ind] = 3
  elif df["Condition"][ind]=='Patchy rain possible':
    df["Condition"][ind] = 4
  else:
    df = df.drop([ind]) #we don't consider other classes so we drop it

#Cast this column to int
df["Condition"] = df["Condition"].astype(int)

df.head() #show the 5 first rows of the dataframe
```

```python
from keras.utils import to_categorical

labels = to_categorical(df.pop('Condition')) #Create classes from the labels

import numpy as np #import numpy library, used for arithmetic

features = np.array(df) #convert our dataframe into ndarray
```

```python
print(features)
print(labels)
```

# Training the model

```python
import sklearn.metrics
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import activations
from tensorflow.keras import layers

NB_classes = 5 #number of outputs
NB_features = 2 #number of inputs
#Split 1 (85% vs 15%)
x_train, x_validate_test, y_train, y_validate_test = train_test_split(features, labels, test_size=0.15, shuffle=True)
#Split 2 (50% vs 50%)
x_test, x_validate, y_test, y_validate = train_test_split(x_validate_test, y_validate_test, test_size=0.50, shuffle=True)
```

```python
model = tf.keras.Sequential()
model.add(layers.Dense(12, activation='relu', input_shape=(NB_features,)))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(NB_classes, activation=tf.keras.activations.softmax))

model.compile(optimizer="adam",loss=tf.keras.losses.categorical_crossentropy, metrics=['accuracy']) #compile the model
model.summary() #to see the param of our model
```

```python
NUM_EPOCHS=20
BATCH_SIZE=10
history = model.fit(x_train, y_train, epochs=NUM_EPOCHS, batch_size=BATCH_SIZE, validation_data=(x_validate, y_validate))
```

```python
performance=model.evaluate(x_test,y_test, batch_size=32, verbose=1, steps=None, )[1] * 100
print('Final accuracy : ', round(performance), '%')
```
